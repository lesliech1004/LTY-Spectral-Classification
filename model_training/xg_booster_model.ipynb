```python
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
import pandas as pd

import model_imports as mi
import os
import numpy as np
```

```python
xgb = XGBRegressor(n_estimators=500,
    max_depth=5,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="reg:squarederror",
    random_state=42,
)
```

```python
xgb.fit(mi.X_train, mi.y_train)
y_pred = xgb.predict(mi.X_test)
```

```python
mi.evaluate_model(model_name="XGB Regressor", model = xgb, y_pred=y_pred, two_bands = True, colors_only = True)
```

```python
#Build a model based off the data set that kept samples with at least 3 bands present.
xgb_3bands = XGBRegressor(n_estimators=500,
    max_depth=5,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="reg:squarederror",
    random_state=42,
)
```

```python
xgb_3bands.fit(mi.X_3bands_train, mi.y_3bands_train)
y_3bands_pred = xgb.predict(mi.X_3bands_test)
```

```python
mi.evaluate_model(model_name="XGB Regressor", model = xgb_3bands, y_pred=y_3bands_pred, two_bands = False, colors_only = True)
```

```python
np.save("y_pred_xgb_2band.npy", y_pred)
np.save("y_pred_xgb_3band.npy", y_3bands_pred)
```
